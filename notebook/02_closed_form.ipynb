```python

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
import time

import sys
sys.path.append('../src')   # âœ… Add your src folder to Python path

from model_closed_form import NormalEquationRegressor
from utils import load_and_preprocess

```

```python
# Load
X_train, X_test, y_train, y_test = load_and_preprocess('../data/processed_data.csv')


```

```python
# Fit closed-form
start = time.time()
model = NormalEquationRegressor()
model.fit(X_train, y_train)
end = time.time()


print('Training time (closed-form):', end-start)


```

```python
# Predict and evaluate
y_pred = model.predict(X_test)
print('MSE:', mean_squared_error(y_test, y_pred))
print('R2:', r2_score(y_test, y_pred))


```

```python
# Save metrics to results
import pandas as pd
metrics = pd.DataFrame([{
'method': 'closed_form',
'mse': mean_squared_error(y_test, y_pred),
'r2': r2_score(y_test, y_pred),
'train_time_s': end-start
}])
metrics.to_csv('../results/metrics_comparison.csv', index=False)
```

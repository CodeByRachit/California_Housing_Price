```python
# 01_data_exploration.ipynb

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.datasets import fetch_california_housing


plt.rcParams['figure.figsize'] = (8,5)
```

```python
housing = fetch_california_housing(as_frame=True)
df = housing.frame
df.head()
```

```python

print(df.shape)
print(df.isnull().sum())
print(df.describe())

```

```python
sns.heatmap(df.corr(), annot=True, fmt='.2f')
plt.title('Feature Correlation')
plt.show()
```

```python

for col in ['MedInc','HouseAge','AveRooms','AveOccup','Latitude','Longitude','MedHouseVal']:
    sns.histplot(df[col], kde=True)
    sns.histplot(df[col], kde=True)
    plt.title(col)
    plt.show()

```

```python
# Optional: create more meaningful features
# e.g., rooms_per_household, population_per_household


df['rooms_per_household'] = df['AveRooms'] / (df['AveOccup'] + 1e-6)
df['population_per_household'] = df['Population'] / (df['AveOccup'] + 1e-6)


# Save processed
df.to_csv('../data/processed_data.csv', index=False)
```
